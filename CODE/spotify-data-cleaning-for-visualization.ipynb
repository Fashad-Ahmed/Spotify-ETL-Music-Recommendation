```python
import pandas as pd
import numpy as np
```

```python
# importing dataset
data = pd.read_csv("../input/spotify-dataset/data/data.csv")
```

```python
# table columns info
data.info()
```

```python
# dropping feature columns:
data = data.drop(['valence', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'mode','speechiness','tempo','key', 'id', 'year'], axis=1)
```

```python
#table info update
data.info()
```

```python
#re-ordering the columns
data = data.reindex(columns=['artists', 'name', 'explicit', 'popularity', 'release_date', 'duration_ms'])
data
```

```python
# turning duration from ms to seconds
data['duration_ms'] = data['duration_ms'].div(1000).round(0).astype(int)
data.rename(columns = {'duration_ms':'duration_s'}, inplace = True)
data['duration_s']
```

```python
#exporting the new dataframe for power BI visualization
data.to_csv('dataset-da.csv')
```
